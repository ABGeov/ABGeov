```python
import pandas as pd
```

```python
df_principal = pd.read_excel("/content/Python-Açoes.xlsx",sheet_name="Principal")
df_principal
```

```python
df_Total_de_acoes = pd.read_excel("/content/Python-Açoes.xlsx",sheet_name="Total_de_acoes")
df_Total_de_acoes
```

```python
df_Ticker = pd.read_excel("/content/Python-Açoes.xlsx",sheet_name="Ticker")
df_Ticker
```

```python
df_ChatGPT = pd.read_excel("/content/Python-Açoes.xlsx",sheet_name="ChatGPT")
df_ChatGPT
```

```python
df_principal = df_principal[['Ativo', 'Data', 'Último (R$)',	'Var. Dia (%)']].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={"Último (R$)":"Valor_final","Var. Dia (%)":"Var_dia_pct"}).copy()
df_principal
```

```python
df_principal["Var_pct"]= df_principal['Var_dia_pct']/100
df_principal["Val_inicial"]=df_principal["Valor_final"]/(df_principal["Var_pct"]+1)
df_principal
```

```python
df_principal=df_principal.merge(df_Total_de_acoes, left_on="Ativo", right_on="Código", how="left" )
df_principal = df_principal.drop(columns=['Código'])
df_principal
```

```python
df_principal = df_principal.drop(columns=['Qtde. Teórica_y'])
df_principal
```

```python
df_principal["Variação_RS"]=(df_principal["Valor_final"]-df_principal["Val_inicial"])*df_principal["Qtde. Teórica_x"]
df_principal
```

```python
pd.options.display.float_format="{:.2f}".format
```

```python
df_principal['Qtde. Teórica_x']=df_principal['Qtde. Teórica_x'].astype(int)
df_principal
```

```python
df_principal=df_principal.rename(columns={"Qtde.Teórica_x":"Qtde_Teorica"}).copy()
df_principal
```

```python
df_principal['Resultado'] = df_principal['Variação_RS'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estável' ))
df_principal
```

```python
df_principal = df_principal.merge(df_Ticker, left_on='Ativo', right_on='Ticker', how='left')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
df_principal = df_principal[['Ativo', 'Data', 'Valor_final', 'Var_dia_pct', 'Var_pct', 'Val_inicial', 'Qtde.Teorica', 'Variação_RS', 'Resultado', 'Nome_y']].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={'Qtde. Teórica':'Qtd_teorica'}).copy()
df_principal = df_principal.rename(columns={'Nome_y':'Nome'}).copy()
df_principal
```
